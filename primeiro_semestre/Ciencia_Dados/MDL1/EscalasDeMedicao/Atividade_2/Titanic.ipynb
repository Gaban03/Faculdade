```python
# prompt: importe o arquivo da sess√£o "passageirosDoTitanic.xslx"

import pandas as pd

# Import the spreadsheet from Google Drive
df = pd.read_csv('passageirosDoTitanic.csv')

# Print the DataFrame
print(df)
```

```python
# prompt: com base no arquivo importado crie um grafico de barras com base nas colunas Sex e Survived (0, 1)

import matplotlib.pyplot as plt

# Create a bar chart with the number of passengers by sex and survival
df.groupby(['Sex', 'Survived']).size().unstack().plot(kind='bar')

# Add labels and title
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.title('Number of Passengers by Sex and Survival')

# Show the plot
plt.show()
```
